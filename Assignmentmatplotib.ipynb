Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib?

matplotlib is a data visualization library in Python used to create static, animated, and interactive visualizations in Python. It provides various kinds of plots, charts, and graphs to represent data in a visually appealing way. Matplotlib is widely used in scientific computing, engineering, data analysis, and machine learning.

Matplotlib is used to create visualizations that help in understanding complex data patterns, relationships, trends, and distributions. It is also used to create publication-quality plots for scientific publications, presentations, and reports.
Pyplot is a module in Matplotlib that provides a convenient interface for creating a variety of plots. Here are five common plots that can be created using the Pyplot module of Matplotlib:
a)Line plot: A line plot is a type of plot that displays data points connected by straight lines. It is used to show trends or changes in data over time.

b)Scatter plot: A scatter plot is a type of plot that displays the relationship between two numerical variables. It is used to visualize patterns, clusters, and outliers in data.

c)Bar plot: A bar plot is a type of plot that displays data using rectangular bars. It is used to compare different categories of data.

d)Histogram: A histogram is a type of plot that displays the distribution of a numerical variable. It is used to visualize the frequency distribution of data.

e)Pie chart: A pie chart is a type of plot that displays data as a circular chart divided into wedges. It is used to show the relative proportions of different categories in a dataset.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

A scatter plot is a type of plot that displays the relationship between two numerical variables. It is used to visualize patterns, clusters, and outliers in data.

```python
import matplotlib.pyplot as plt 
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('my title')
plt.show()

```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])?

The subplot() function is used to create multiple plots in a single figure. It takes three arguments - the number of rows, the number of columns, and the index of the subplot. By specifying the index of each subplot, you can arrange the plots in a grid.



```python

import numpy as np
import matplotlib.pyplot as plt

# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure with four subplots
fig, axs = plt.subplots(2, 2)

# Plot line 1 in the first subplot
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

# Plot line 2 in the second subplot
axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

# Plot line 3 in the third subplot
axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

# Plot line 4 in the fourth subplot
axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

# Adjust the spacing between subplots
plt.subplots_adjust(wspace=0.3, hspace=0.5)

# Show the plot
plt.show()
```

A bar plot is a graphical representation of data that uses rectangular bars to represent the values of each category or group of data. The length or height of each bar represents the value of the corresponding data category or group.

Bar plots are commonly used to display and compare the values of different categories or groups of data. They are particularly useful for visualizing and analyzing discrete or categorical data. Bar plots are also useful for displaying data that is not continuous or that cannot be easily measured on a continuous scale.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit)
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

 box plot is a graphical representation of data that displays the distribution of a dataset using five statistics: the minimum value, the first quartile (Q1), the median (Q2), the third quartile (Q3), and the maximum value.

Box plots are useful for summarizing the distribution of a dataset and identifying potential outliers. They are particularly useful for comparing the distributions of different datasets or groups of data.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create a box plot
data = [box1, box2]
plt.boxplot(data)
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.xlabel('Box')
plt.ylabel('Value')
plt.title('Box Plot')
plt.show()
```
